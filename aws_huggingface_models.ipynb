```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
classifier("i want that badly")
```

```python
results=classifier(["I tried to call xyz  but i don't think she is interested in me anymore.","i also tried to talk abc he like it but he resist to work with me i don't know why he want to he with me."])
for result in results:
  print(f"label:{result['label']},with score:{round(result['score'], 4)}")
```
